# Data Visualization

```python
# Distribution of satisfaction levels
plt.figure(figsize=(7,5))
sns.countplot(x="satisfaction",data=df, palette='mako')
plt.title("Distribution of satisfaction levels")
plt.show()
```

```python
# Age vs Satisfaction
plt.figure(figsize=(8,5))
sns.boxplot(x="Age",y="satisfaction",data=df, palette='rocket')
plt.title("Age vs Satisfaction")
plt.show()
```

```python
#Departure Delay vs Arrival Delay by Satisfaction
plt.figure(figsize=(8,5))
sns.scatterplot(x="Departure Delay in Minutes",y="Arrival Delay in Minutes",data=df,hue="satisfaction")
plt.title("Departure Delay vs Arrival Delay by Satisfaction")
plt.show()
```

```python
numeric_data = df.select_dtypes(include=['number'])
```

```python
#Correlation Heatmap (Numeric Columns Only)
plt.figure(figsize=(12, 10))
sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
#Gender vs Satisfaction
plt.figure(figsize=(8,5))
sns.countplot(x="Gender",hue="satisfaction",data=df, palette='flare')
plt.title("Gender vs Satisfaction")
plt.show()
```

```python
#Customer Type vs Satisfaction
plt.figure(figsize=(8,5))
sns.countplot(x="Customer Type",hue="satisfaction",data=df, palette='magma')
plt.title("Customer Type vs Satisfaction")
plt.show()
```

```python
#Inflight Wifi Service Ratings
plt.figure(figsize=(8, 5))
sns.barplot(x=df['Inflight wifi service'].value_counts().index, y=df['Inflight wifi service'].value_counts().values, palette='Blues')
plt.title('Inflight Wifi Service Ratings')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()
```

```python
#Flight Distance Distribution by Class
plt.figure(figsize=(8, 5))
sns.kdeplot(x='Flight Distance', hue='Class',data=df)
plt.title('Flight Distance Distribution by Class')
plt.show()
```

```python
#Age Distribution by Gender
plt.figure(figsize=(8,5))
sns.histplot(data=df, x='Age', hue='Gender', palette='rocket')
plt.title('Age Distribution by Gender')
plt.show()
```

```python
#Flight Distance Distribution by Satisfaction
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Flight Distance', hue='satisfaction', kde=True,palette='cubehelix')
plt.title('Flight Distance Distribution by Satisfaction')
plt.show()
```

```python
#Flight Distance vs. Age
plt.figure(figsize=(8,5))
sns.lineplot(data=df, x='Age', y='Flight Distance')
plt.title('Flight Distance vs. Age')
plt.xlabel('Age')
plt.ylabel('Flight Distance')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.pairplot(data=numeric_data, vars=['Age', 'Flight Distance', 'Arrival Delay in Minutes'], palette='seagreen')
plt.show()
```

```python
# Gender vs Inflight Wifi Service Ratings
plt.figure(figsize=(8, 5))
sns.barplot(x='Gender', y='Inflight wifi service', data=df, palette='muted')
plt.title('Gender vs Inflight Wifi Service Ratings')
plt.show()
```

# Insights and Recommendations
1. Most passengers report being either satisfied or neutral, with fewer passengers reporting dissatisfaction.
2. Older passengers generally report higher satisfaction compared to younger passengers.
3. There is a strong positive correlation between departure delays and arrival delays.
4. Inflight service, cleanliness, and comfort have the highest positive correlation with satisfaction.
5. Both male and female passengers report similar satisfaction levels.
6. Loyal customers report significantly higher satisfaction levels compared to first-time flyers.
7. Ratings for inflight wifi service are generally low, indicating dissatisfaction with this service.
8. Business Class passengers tend to take longer flights compared to Economy Class passengers.
9. The age distribution shows a balanced mix of genders across various age groups.
10. Passengers traveling shorter distances report higher satisfaction levels.
11. Older passengers tend to take longer flights.
12. There is no significant difference in inflight wifi service ratings between genders.
