# Pandas

* Pandas is a data manipulation and data preprocessing library
* Comes with built in functions to handle data in many ways
* Multiple options for connections made pandas first choise for data extractions
* Handling inconsistant as well noisy data is very easy in pandas
* Pandas has two data structures - 1) Series and 2) DataFrame
* Series is used to store 1D data and DataFrame can be used to store 2D data
* Series in pandas is the Indexed Series helps to perform wide operations on table columns
* Every column of DataFrame (Table) in pandas is a Series itself
* Pandas works very closely with numpy, makes it faster in data operations
