```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv(r"C:\Users\Soft\Desktop\Py\Docs\Salary Data\Salary_dataset.csv")
```

```python
df.head()
```

```python
df.drop("Unnamed: 0", axis=1, inplace=True)
```

```python
plt.figure(figsize=(10, 4))
plt.scatter(df["YearsExperience"], df["Salary"])
plt.xlabel("Year of Experience")
plt.ylabel("Salary")
plt.title("Experience vs Salary")
plt.grid()
plt.show()
```

```python
X = df["YearsExperience"]
y = df["Salary"]
```

```python
m = np.sum((X-X.mean())*(y-y.mean()))/np.sum(np.square(X-X.mean()))
m
```

```python
c = y.mean() - (m * X.mean())
c
```

```python
y_pred = m * X + c
```

```python
plt.figure(figsize=(10, 4))
plt.scatter(df["YearsExperience"], df["Salary"])
plt.plot(df["YearsExperience"], y_pred, color="orange")
plt.xlabel("Year of Experience")
plt.ylabel("Salary")
plt.title("Experience vs Salary")
plt.grid()
plt.show()
```

```python
exp = 5
sal = m * exp + c
f"Experience {exp} years, Expected Salary is {np.round(sal)}"
```

```python
plt.figure(figsize=(10, 4))
plt.scatter(df["YearsExperience"], df["Salary"])
plt.plot(df["YearsExperience"], y_pred, color="orange")
plt.scatter(exp, sal, color="red")
plt.xlabel("Year of Experience")
plt.ylabel("Salary")
plt.title("Experience vs Salary")
plt.grid()
plt.show()
```

```python
SSR = np.sum(np.square(y_pred - y))
SSR
```

```python
MSE = np.mean(np.square(y_pred - y))
MSE
```

```python
RMSE = np.sqrt(MSE)
RMSE
```
